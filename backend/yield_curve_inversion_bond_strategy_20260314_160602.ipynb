```python
def generate_signals(macro, prices):
    signals = pd.DataFrame(0, index=macro.index, columns=prices.columns)
    # Long TLT when yield curve 10y2y spread inverted
    signals['TLT'] = (macro['spread_10y2y'] < 0).astype(float)
    signals = signals.shift(1)
    signals = signals.resample('ME').last()
    # Align to daily prices index
    signals = signals.reindex(prices.index, method='ffill').fillna(0)
    return signals
```
